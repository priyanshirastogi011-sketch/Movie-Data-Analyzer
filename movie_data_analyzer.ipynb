```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("movies.csv")
print(df.head())
print(df.info())
print(df.describe())

df.dropna(inplace=True)  #remove missing values
df.drop_duplicates(inplace=True)   #remove duplicate rows
```

```python
highest = df[df["IMDB_Rating"] == df["IMDB_Rating"].max()]        #Highest Rated Movie
print(highest)
```

```python
lowest = df[df["IMDB_Rating"] == df["IMDB_Rating"].min()]        #Lowest Rated Movie
print(lowest)
```

```python
average = df["IMDB_Rating"].mean()     #Average Rating
print(average)
```

```python
total_movies = df.shape[0]   #total movies 
print(total_movies)
```

```python
movies_per_year = df["Released_Year"].value_counts().sort_index()  #Movies Released Each Year
print(movies_per_year)
```

```python
most_common_genre = df["Genre"].value_counts()
print(most_common_genre)
```

```python
print(df["Runtime"].dtype)     #Average Runtime
df["Runtime"].mean()
```

```python
top_10_movies = df.sort_values("IMDB_Rating", ascending=False)[
    ["Series_Title", "IMDB_Rating", "Released_Year", "Genre"]
].head(10)                                                     #Top 10 Movies

print(top_10_movies)
```

```python
worst_10_movies = df.sort_values("IMDB_Rating", ascending=False).tail(10)     #Worst 10 Movies
print(worst_10_movies)
```

```python
print(df["Released_Year"].dtype)
df["Released_Year"] = pd.to_numeric(df["Released_Year"], errors="coerce")
```

```python
movies_after_2015 = df[df["Released_Year"] > 2015][
    ["Series_Title", "Released_Year", "IMDB_Rating", "Genre"]     
]                                                                 #Movies Released After 2015

print(movies_after_2015)
```

```python
movies_per_year = df["Released_Year"].value_counts().sort_index()

plt.bar(movies_per_year.index, movies_per_year.values)
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.title("Movies Released Every Year")
plt.xticks(rotation=45)

plt.savefig("output/movies_per_year.png")
plt.show()
```

```python
top_genres = df["Genre"].value_counts().head(10)

plt.bar(top_genres.index, top_genres.values)
plt.xlabel("Genres")
plt.ylabel("Number of Movies")
plt.title("Top 10 Genres")
plt.xticks(rotation=45)

plt.savefig("output/top_genres.png")
plt.show()
```

```python
#Ratings Distribution Histogram

plt.figure(figsize=(8,5))

plt.hist(df["IMDB_Rating"], bins=np.arange(7.5, 9.6, 0.1))

plt.title("Ratings Distribution")
plt.xlabel("IMDb Rating")
plt.ylabel("Number of Movies")

plt.grid(axis='y', alpha=0.3)

plt.savefig("output/ratings_histogram.png")
plt.show()
```

```python
# Top 10 movies by IMDb rating
top10 = df.sort_values(by="IMDB_Rating", ascending=False).head(10)

plt.figure(figsize=(10,6))

plt.barh(top10["Series_Title"], top10["IMDB_Rating"])

plt.title("Top 10 Movies by IMDb Rating")
plt.xlabel("IMDb Rating")
plt.ylabel("Movie Title")

plt.gca().invert_yaxis()  # Highest rated movie on top

plt.tight_layout()

plt.savefig("output/top10_movies.png")
plt.show()
```

```python
#Runtime Distribution Histogram
plt.figure(figsize=(8,5))

plt.hist(df["Runtime"], bins=15)

plt.title("Runtime Distribution of Movies")
plt.xlabel("Runtime (Minutes)")
plt.ylabel("Number of Movies")

plt.grid(axis='y', alpha=0.3)

plt.savefig("output/runtime_histogram.png")
plt.show()
```

```python
# Top 10 movies by IMDb rating
top10 = df.sort_values(by="IMDB_Rating", ascending=False).head(10)

# Select required columns
report = top10[["Series_Title", "IMDB_Rating", "Genre"]]

# Save to CSV
report.to_csv("Top10Movies.csv", index=False)

print("Top10Movies.csv created successfully!")
```

```python
# Total movies
total_movies = len(df)

# Highest rated movie
highest_rated_movie = df.loc[df["IMDB_Rating"].idxmax(), "Series_Title"]

# Average rating
average_rating = round(df["IMDB_Rating"].mean(), 1)

# Most common genre
most_common_genre = (
    df["Genre"]
    .str.split(", ")
    .explode()
    .mode()[0]
)

# Average runtime
average_runtime = round(
    df["Runtime"].str.replace(" min", "", regex=False).astype(int).mean()
)

# Top movie
top_movie = highest_rated_movie

# Print summary
print("=" * 42)
print("       Movie Data Analyzer")
print("=" * 42)
print(f"Total Movies : {total_movies}")
print()
print(f"Highest Rated Movie : {highest_rated_movie}")
print()
print(f"Average Rating : {average_rating}")
print()
print(f"Most Common Genre : {most_common_genre}")
print()
print(f"Average Runtime : {average_runtime} minutes")
print()
print("Top Movie :")
print(top_movie)
print()
print("Analysis Completed Successfully!")
print("=" * 42)
```
